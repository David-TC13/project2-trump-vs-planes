```python
import pandas as pd
from pandas import json_normalize

import requests 
import json
import os
from dotenv import load_dotenv 

import time
```

```python
load_dotenv()
```

```python
key= os.getenv("token")
```

```python
url= f"https://applications.icao.int/dataservices/api/incident?api_key={key}&StateOfOccurrence=&format=&StateOfOperator=&StateOfRegistry=&Year=2020"
```

```python
res = requests.get(url)
```

```python
plane_crash= res.json()
```

```python
df_plane= pd.DataFrame(json_normalize(plane_crash))
```

```python
df_plane.dropna(how= 'all',inplace= True)
```
